We now convert the `hospitalCases` column to a numpy array `yvals`

## 2a) Plot the data `yvals` vs `xvals` in a scatterplot

```python
plt.scatter(xvals, yvals, color='blue', label='hospitalCases')
plt.title('Hospital Cases')
plt.xlabel('Days since 2020-03-01')
plt.ylabel('Number of hospital cases')
plt.legend()
plt.show()
```

## 2b) Fit an exponential model to the data 

From our knowledge of how the virus spreads, we know that the number of infections, hospital admissions etc. should (roughly) follow an exponential curve. We would therefore like to fit a model of the form $y = c_1 e^{c_2 x}$, where $y$ is the number of admissions on day $x$. 

Note that this is a linear model on a log-scale for $y$. That is, $\log y = \log c_1 +  c_2 x$.

- Fit a linear model for $\log(\text{yvals})$ vs $\text{xvals}$, and print the values of $c_1$ and $c_2$
- Plot the fit $y = c_1 e^{c_2x}$ along with the scatterplot of the data

```python
# your code here
def expofit(y, x):
    """
    Fit an exponential function to the data
    :param y: y values
    :param x: x values
    """
    Y = np.log(y)
    X = np.array([np.ones(len(x)),x]).T  # Create a 2D array with x and a column of ones
    beta_values = np.linalg.lstsq(X, Y, rcond=None)[0]  # Solve the least squares problem
    return beta_values  # Return the slope and intercept of the fitted line

beta_values = expofit(yvals, xvals)  # Call the function with yvals and xvals as arguments
plt.scatter(xvals, yvals, color = 'blue', label='hospitalCases')  # Plot the original data points
plt.plot(xvals, np.exp(beta_values[1] * xvals + beta_values[0]), 'r-', label='Exponential Fit')  # Plot the fitted line
plt.title('Exponential Fit to Hospital Cases')
plt.xlabel('Days since 2020-03-01')
plt.ylabel('Number of hospital cases')
plt.legend()
plt.show()  # Show the plot
```
